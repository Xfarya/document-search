```python
import pandas as pd
import random
import numpy as np
```

```python
# Load the Parquet file into a pandas DataFrame

df = pd.read_parquet("test-00000-of-00001.parquet", engine='pyarrow')
df.head()
```

```python
# Create a list of randomised neg documents per query; there are 9650 queries (rows)

c = -1
df['neg_sampling'] = [0] * len(df)
df['triplets'] = [0] * len(df)

for x in df['query']:
    c += 1
    
    # positive_doc array below (includes the human-scored 'most relevant' document as a 1, other relevant docs as 0)
    most_relevant_nums = df['passages'][c]['is_selected']
    # print(most_relevant_nums)

    # positive_doc titles below
    relevant_docs = df['passages'][c]['passage_text']
    # print(most_relevant_chars)

    # initialising variables for finding documents irrelevant to a particular query
    rand_ints = []
    rand_docs = []
    # generating random query rows to select irrelevant docs from
    while len(rand_ints) < len(most_relevant_nums):
        nums = random.randint(0, len(df)-1)
        if nums not in rand_ints and nums != c:
            rand_ints.append(nums)
    # selecting random irrelevant documents
    for x in rand_ints:
        story = np.random.choice(df['passages'][x]['passage_text'])
        rand_docs.append(story)
    # creating a column per query which contains random, irrelevant documents
    df['neg_sampling'][c] = rand_docs
    # the df now contains a neg_sampling column with a series of irrelevant documents per query row

    # Now we will generate triplets from a row's query, relevant docs and irrelevant docs
    triplet_list = []
    triplet_counter = 0
    while len(triplet_list) < len(most_relevant_nums):
        triplet_list.append((df['query'][c], relevant_docs[triplet_counter], rand_docs[triplet_counter]))
        triplet_counter += 1
    df['triplets'][c] = triplet_list

    # Now we have a list of triplets per query.
df['triplets'][1] #Example triplets for a query 
# df 
```
